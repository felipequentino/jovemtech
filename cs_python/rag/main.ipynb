```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage

from dotenv import load_dotenv
load_dotenv()
```

```python
# Inicializar o modelo
llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0.7
)

pergunta = "a sabrina carpenter levou algum prêmio no grammy ? "

message = [
    HumanMessage(content=pergunta)
]

resposta = llm.invoke(message)
print(f"Pergunta: {pergunta}")
print(f"Resposta: {resposta.content}")
```

```python
from langchain_community.tools import DuckDuckGoSearchResults

search = DuckDuckGoSearchResults(num_results=3)

pergunta = "a sabrina carpenter levou algum prêmio no grammy ? "

# Buscar contexto na web
print("Buscando na web...")
contexto = search.invoke(pergunta)
print(f"Contexto encontrado: {contexto[:200]}...\n")

prompt = f"Contexto da web: {contexto}\n\nPergunta: {pergunta}"
resposta = llm.invoke(prompt)

print(f"Resposta: {resposta.content}")
```
